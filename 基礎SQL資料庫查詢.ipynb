# SELECT 計算一個衍生變數
## 人均 GNP (GNP per Capita) 

```
dbname = "world"
conn = create_engine('mysql+mysqldb://{}:{}@{}:{}/{}'.format(user, passwd, host, port, dbname))
query_str = """
SELECT Name,
    GNP / Population
  FROM country;
"""
GNP_per_capita = pd.read_sql(query_str, conn)
GNP_per_capita
```

# AS 
## 為衍生變數取名:

```
dbname = "world"
conn = create_engine('mysql+mysqldb://{}:{}@{}:{}/{}'.format(user, passwd, host, port, dbname))
query_str = """
SELECT Name,
    GNP / Population AS gnp_percap
  FROM country;
"""
GNP_per_cap = pd.read_sql(query_str, conn)
GNP_per_cap
```

# Q : 隨堂練習 : 將 nba.nba_players 中球員的身高 heightMeters 換算為公分並取名為 heightCentimeters

```
dbname = "nba"
conn = create_engine('mysql+mysqldb://{}:{}@{}:{}/{}'.format(user, passwd, host, port, dbname))
query_str = """
SELECT personName,
    heightMeters,
    heightMeters * 100 AS heightCentimeters
  FROM nba_players;
"""
height_mtocm = pd.read_sql(query_str, conn)
height_mtocm
```

# 使用函數 : 
## ROUND(, decimal) 函數可以調整小數點位數

```
dbname = "world"
conn = create_engine('mysql+mysqldb://{}:{}@{}:{}/{}'.format(user, passwd, host, port, dbname))
query_str = """
SELECT Name,
    ROUND(GNP / Population, 2) AS gnp_percap 
  FROM country;
"""
gnp_percap = pd.read_sql(query_str, conn)
gnp_percap
```

# Q : 隨堂練習使用 POWER(, exponent) 函數計算 nba.nba_players 中球員的 bmi

```
dbname = "nba"
conn = create_engine('mysql+mysqldb://{}:{}@{}:{}/{}'.format(user, passwd, host, port, dbname))
query_str = """
SELECT personName,
    heightMeters,
    weightKilograms,
    ROUND(weightKilograms / POWER(heightMeters,2), 2) AS bmi 
  FROM nba_players;
"""
bmi = pd.read_sql(query_str, conn)
bmi
```
